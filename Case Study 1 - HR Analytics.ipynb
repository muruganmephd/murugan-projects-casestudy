```python
a = ['dept', 'numberOfProjects', 'timeSpent.company', 'workAccident', 'promotionInLast5years', 'salary']
```

```python
a
```

```python
fig  = plt.subplots(figsize=(20,25))
for i,j in enumerate(a): 
    print(i,j)
    plt.subplot(3,2,i+1)
    sns.countplot(x=j, data=df, hue='left') 
    plt.xticks(rotation = 90)
plt.show()
```

```python
#Conclusions:-

#Dept graph : Inference => Sales, Support, Technical departments having high attrition rate
#             Suggestion-> Consider to give Increment, Incentives

#Projects graph : Inference -> Employees are leaving when they either have highest or lowest no.of projects
#                 Suggestion -> 3 (or) 4 no. of projects are safest that company should give to an employee
    
#Time spent graph : Inference -> 3-5 years of experience is the most crucial time for the company
#                   Suggestion -> Consider to give Annual Incentives
    
#Work Accident graph : Inference -> Most of the employees who are leaving, are not invloved in any work accident
#                      Suggestion -> Work accident should be avoided. However, work accident is not the main reason for 
#                                    attrition

#Promotion in the last 5 years graph : Inference -> Can be biggest reasons, as almost all the employees are leaving, 
#                                                   haven't been promoted
#                                      Suggestion -> Consider to give promotions for deserving employees periodically. 

#Salary graph : Inference -> Most of employees who are leaving are getting low/medium salary.
#               Suggestion -> Consider to give competitive salary for asset employees.
```
