```python
# Drop the catagorical column from dataset

df_drop = df.copy()  # form the new dataset df_drop without catagorical column
player = df_drop.pop('Player')   # pop out the player column
```

```python
# some columns have very big values and some columns have small values. so standardization is required for the dataset.
```

```python
#Rescaling the features
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_drop)
df_scaled
```

```python

df_df1 = pd.DataFrame(df_scaled,columns = [ 'Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100','50', '0', 'Exp'])
df_df1.head()
```

```python
# Build the clusters

from sklearn.cluster import KMeans
```

```python
#Elbow curve to find the ideal number of clusters.
#ssd (sum of squared distance) - inertia

clusters = list(range(2,9))          # always minimum number of clusters is 2 and maximum no of clusters is random
ssd = []
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters, max_iter=150 )     # maximum iteration taken as randomly
    model_clus.fit(df_df1)
    ssd.append(model_clus.inertia_)
```

```python
# plot the elbow curve
plt.plot(clusters,ssd)
plt.ylabel("SSD")
```

```python
# Here, elbow sharp points indicates 3 and 4. so no of clusters taken as either 3 or 4.
# Here no of clusters taken as 4
```

```python
cluster = KMeans(n_clusters=4, max_iter=150, random_state= 15)
cluster.fit(df_df1)
```

```python
#Cluster labels

cluster.labels_
```

```python
# Assign the label in original dataset

df['Cluster_Id'] = cluster.labels_
df.head()
```
