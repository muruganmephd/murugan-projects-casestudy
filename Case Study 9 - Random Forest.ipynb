```python
churn_data = churn_data.dropna()
```

```python
# Here zero null values
```

```python
churn_data.columns
```

```python
# As i can see that customer ID is not the important column to predict my customer is churning out or not so we will directly 
# drop it.
```

```python
churn_data.drop(columns=['customerID'], inplace=True)
```

```python
churn_data["PhoneService"].value_counts()
```

```python
churn_data["Contract"].value_counts()
```

```python
#label encoding - to convert object data into numerical
col_list = [] #to add all the object columns into this list

for col in churn_data.columns:
    if(churn_data[col].dtype == "object"):
        col_list.append(col)
print(col_list)
```

```python
# we don't need to label encode the Churn column. so pop out the churn column from the list.
```

```python
col_list.pop(-1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
for i in col_list:
    churn_data[i] = label_encoder.fit_transform(churn_data[i])
```

```python
churn_data.head(20)
```

```python
churn_data["Contract"].value_counts()
```

```python
#Dividing the data into X and Y - Independent and Dependent feature
X = churn_data.iloc[:,:-1] #features or Independent variables
y = churn_data.iloc[:,-1] #target data
```

```python
#Splitting the data to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 44)
```

```python
#Building the Random forest model
classifier1 = RandomForestClassifier(n_estimators=400) #No.of decision trees = 400
classifier1.fit(X_train, y_train) #traning the model
```

```python
y_pred = classifier1.predict(X_test) #predictions for testing data
```

```python
y_pred
```

```python
# check the accuracy
a1 = accuracy_score(y_test, y_pred)
a1
```

```python
n = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
accuracy_l = []

for i in n:
    Rd_model = RandomForestClassifier(n_estimators=i)
    Rd_model.fit(X_train, y_train)
    pred = Rd_model.predict(X_test)
    a = accuracy_score(y_test, pred) * 100
    accuracy_l.append(a)
accuracy_l
```

```python
# To make the dataframes for no. of trees and accuracy
d = {"Number of Trees":n,"Accuracy": accuracy_l}
df = pd.DataFrame(d)
print(df)
```
