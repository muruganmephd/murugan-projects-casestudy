```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Checking the null values
df.isnull().sum()
```

```python
# If null values are in single or double or three columns, drop the null values.
# If null values are spread in more columns, then replace the null values with other values.
```

```python
# drop the null values

df.dropna(inplace=True)
```

```python
# Checking the duplicates
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# Now the 23 duplicate rows are deleted.
```

```python
# Rename the unnamed column
df.rename(columns={'Unnamed: 14':'Income'},inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

```python
# Checking the outliers

for col in df.columns:
  if df[col].dtype!="object":           # if column is not object
    plt.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
# Removal of outliers
# Here, capital-gain and capital-loss columns are not having boxplot. so omit those columns.

out_list=['age','fnlwgt','education-num','hours-per-week']
```

```python
out_list
```

```python
for col in out_list:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1
  UW=Q3+1.5*IQR             # UW = UNDER WHISKER, LW = LOWER WHISKER
  LW=Q1-1.5*IQR
  df=df[(df[col]>=LW) & (df[col]<=UW)]
```

```python
for col in out_list:
  plt.boxplot(df[col])
  plt.xlabel(col)
  plt.show()
```

```python
df
```

```python
# Almost more than 9000 rows are removed.
```

```python
le=LabelEncoder()
```

```python
for col in df.columns:
  if df[col].dtype=="object":
    df[col]=le.fit_transform(df[col])
```
